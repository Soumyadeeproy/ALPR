```python
def calculate_iou(sx,sy,ex,ey,sx1,sy1,ex1,ey1):
    i_s_x = max(sx,sx1)
    i_s_y = max(sy,sy1)
    i_e_x = min(ex,ex1)
    i_e_y = min(ey,ey1)
    
    ia = max(i_e_x - i_s_x, 0) * max(i_e_y - i_s_y, 0)
    a1  = (ex  - sx)  * (ey  - sy)
    a2  = (ex1 - sx1) * (ey1 - sy1)
    ua = a1 + a2 - ia
    
    return ia / float(ua)
```

```python
import os
import h5py 
import numpy as np
import pandas as pd

anchors = [(12,12,2,2),(19,20,3,4),(22,22,5,5),(34,35,6,7),(42,42,8,8),(49,50,9,10)]

feature_folder = 'FEATURE_MAPS'
anchor_folder = 'ANCHORS'

image_details = pd.read_csv('DATASET/DETAILS.csv')
anchor_details = {'anchor_name': [], 'iou': []}

for _, row in image_details.iterrows():
    name = row['image_name']
    print(name)
    feature_map = h5py.File(feature_folder + "/" + name.split('.')[0] + '.h5','r+')
    feature_map = np.array(feature_map.get('default'))
    [ n , n_i, n_j, n_k ] = feature_map.shape
    
    true_start_x = row['start_x']
    true_start_y = row['start_y']
    true_end_x = row['end_x']
    true_end_y = row['end_y']

    cnt = 1    
    
    for w_l, w_r, h_l, h_r in anchors:
        for i in range( h_l, n_i-h_r, 10):
            for j in range( w_l, n_j-w_r, 10):
                top_left_x = i - h_l
                top_left_y = j - w_l
                bottom_right_x = i + h_r
                bottom_right_y = j + w_r
                
                #output_name = fm.split('.')[0]+'_anchor_'+str(cnt)+'.h5'
                
                start_x = (4 * top_left_x) - 16
                start_y = (4 * top_left_y) - 16
                end_x = (4 * bottom_right_x) + 16
                end_y = (4 * bottom_right_y) + 16
                
                start_x = max(0, start_x)
                start_y = max(0, start_y)
                end_x = min(4*n_i, end_x)
                end_y = min(4*n_j, end_y)
                
                iou = calculate_iou(true_start_x, true_start_y, true_end_x, true_end_y, start_x, start_y, end_x, end_y)
                
                anchor_details['anchor_name'].append(name.split('.')[0]+'_anchor_'+str(cnt)+'.h5')
                anchor_details['iou'].append(iou)
                
                cnt += 1

df = pd.DataFrame(anchor_details)
df.to_csv('ANCHOR_DETAILS.csv')
print('Done')
```
