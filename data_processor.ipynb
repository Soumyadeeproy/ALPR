```python
import os
import h5py
import cv2

with h5py.File('DATASET/images.h5', 'w') as f:
    for im in os.listdir('DATASET'):
        if 'image' in im and '.jpg' in im:
            print(im)
            image = cv2.imread(os.path.join('DATASET', im))
            image = image.reshape(1, *image.shape)
            f.create_dataset(im, data=image, compression="gzip")
```
